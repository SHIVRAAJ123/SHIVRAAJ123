```python
import os
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils import shuffle

# For visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Parameters
img_height, img_width = 224, 224
batch_size = 32
train_data_dir = '/content/drive/My Drive/datasets/train_data'  # Replace with your training dataset path
val_data_dir = '/content/drive/My Drive/datasets/val_data'      # Replace with your validation dataset path

# Image Data Generators
train_datagen = ImageDataGenerator(rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
    )

validation_generator = val_datagen.flow_from_directory(
    val_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode ='binary'
    )

# Create data generators without augmentation
train_generator_no_aug = ImageDataGenerator(rescale=1./255).flow_from_directory(
    train_data_dir,
    target_size=(224, 224),
    batch_size=1,  # Set batch size to 1 to get each image individually
    class_mode='binary',
    shuffle=False
)

# Extract data and labels
train_data, train_labels = next(zip(*[(train_generator_no_aug[i][0], train_generator_no_aug[i][1]) for i in range(len(train_generator_no_aug))]))

# Resample
ros = RandomOverSampler()
train_data_resampled, train_labels_resampled = ros.fit_resample(train_data, train_labels)

# Shuffle the resampled data
train_data_resampled, train_labels_resampled = shuffle(train_data_resampled, train_labels_resampled)

# Use the resampled data to create a new data generator
train_datagen_resampled = ImageDataGenerator(rescale=1./255)
train_generator_resampled = train_datagen_resampled.flow(train_data_resampled, train_labels_resampled, batch_size=32)

```

```python
# Check class distribution
train_class_counts = train_generator.classes
val_class_counts = validation_generator.classes

print(f'Training class distribution: {np.bincount(train_class_counts)}')
print(f'Validation class distribution: {np.bincount(val_class_counts)}')

# Optionally, use class weights to handle imbalance
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_class_counts),
    y=train_class_counts
)

class_weight_dict = dict(enumerate(class_weights))
print(f'Class weights: {class_weight_dict}')
```

```python
# Load ResNet-50 model without the top layer
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Adding custom layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

# Defining the model
model = Model(inputs=base_model.input, outputs=predictions)

# Freezing the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_steps=validation_generator.samples // batch_size,
    validation_data=validation_generator,
    epochs=10
)

# Save the trained model
model.save('/content/drive/My Drive/currency_detection_model1.h5')
```

```python
# Load the saved model
model = tf.keras.models.load_model('/content/drive/My Drive/currency_detection_model1.h5')

# Evaluate the model on the validation set
# Calculate the number of steps per epoch correctly
val_loss, val_accuracy = model.evaluate(validation_generator, steps=validation_generator.samples // batch_size + 1)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")

# Predict classes
validation_generator.reset()
Y_pred = model.predict(validation_generator, validation_generator.samples // batch_size + 1)
y_pred = (Y_pred > 0.5).astype(int).reshape(-1)

# Get the true labels
y_true = validation_generator.classes

# Classification report
print(classification_report(y_true, y_pred, target_names=['Not Currency', 'Currency'])) # Change target_names to a single name

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Currency', 'Currency'], yticklabels=['Not Currency', 'Currency']) # Change xticklabels and yticklabels to a single name
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Directory containing your photos
photos_dir = '/content/drive/My Drive/Money'

# Function to preprocess and predict
def predict_image(img_path):
    # Import the image module from keras.preprocessing
    from keras.preprocessing import image
    # Load the image
    img = image.load_img(img_path, target_size=(224, 224))

    # Convert the image to array
    img_array = image.img_to_array(img)

    # Expand dimensions to match the input shape of the model
    img_array = np.expand_dims(img_array, axis=0)

    # Re hubscale the image
    img_array /= 255.0

    # Predict the class
    prediction = model.predict(img_array)

    # Convert prediction to class label
    class_label = 'Counterfeit' if prediction[0] < 0.5 else 'Genuine'

    # Plot the image with prediction
    plt.imshow(img)
    plt.title(f'Prediction: {class_label}')
    plt.axis('off')
    plt.show()

# List all images in the directory
images = os.listdir(photos_dir)

# Predict each image
for img_name in images:
    img_path = os.path.join(photos_dir, img_name)
    predict_image(img_path)
```
